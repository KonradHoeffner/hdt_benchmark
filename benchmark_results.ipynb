# Bencharking Sophia

This benchmark compares the performances of the [sophia](https://github.com/pchampin/sophia_rs) library
with other RDF libraries.

It was executed on an Intel(R) Core(TM) i7-8550U CPU @ 1.80GHz, with 8 GiB of RAM.

```python
%matplotlib inline
from plot_utils import *
```

## The `query` test

This test consists in loading an N-Triple file in memory,
and querying all the triples with a given subject and object.

It has been tested with the following tools:

* `jena`: [Apache Jena](http://jena.apache.org/) 3.10.0 compiled and executed with OpenJDK 11.0.1
* `n3js`: [N3.js](https://github.com/rdfjs/N3.js#readme) 1.0.0-beta.1 executed with Node.js v8.11.4
* `librdf`: [librdf](http://librdf.org/) 1.0.17 compiled with clang 7.0.0-3
* `pypy`: [RDFlib](http://rdflib.readthedocs.org/) 4.2.2 executed with PyPy 5.8.0-beta2
* `python`: [RDFlib](http://rdflib.readthedocs.org/) 4.2.2 executed with Python 3.5.3
* `sophia`: [sophia](https://github.com/pchampin/sophia_rs) v0.1.0 compiled with rustc 1.32.0
  (using [FastGraph](https://docs.rs/sophia/latest/sophia/graph/inmem/type.FastGraph.html))
* `sophia_lg`: [sophia](https://github.com/pchampin/sophia_rs) v0.1.0 compiled with rustc 1.32.0
  (using [LightGraph](https://docs.rs/sophia/latest/sophia/graph/inmem/type.LightGraph.html))


```python
plot_query_stats(load_data("query", "sophia", "sophia_lg", "librdf", "jena", "n3js", "python", "pypy"))
```

## The `parse` test

This test consists in parsing an N-Triple file without storing the triples.
The purpose is to evaluate the performances of the *parser*.

```python
data = load_data("parse", "sophia", "librdf", "n3js")
my_plot(data, "t_parse", loglog=1, title="Time (in s) to parse an NT file")
```
